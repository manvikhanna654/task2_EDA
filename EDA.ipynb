```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load Titanic dataset
df = sns.load_dataset('titanic')  # If you're using a CSV: pd.read_csv('your_path.csv')

# Set plot style
sns.set(style="whitegrid")

# Basic info
print("Dataset Shape:", df.shape)
print(df.info())
print(df.describe(include='all'))

# Show first few rows
print(df.head())
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values in Dataset")
plt.show()
numerical_cols = ['age', 'fare']

# Histograms
df[numerical_cols].hist(bins=30, figsize=(10, 5), layout=(1, 2))
plt.suptitle("Histograms of Age and Fare")
plt.show()

# Boxplots
plt.figure(figsize=(10, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 2, i+1)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
# Pairplot
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived')
plt.show()

# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
# Survival count by sex
sns.countplot(data=df, x='sex', hue='survived')
plt.title("Survival by Sex")
plt.show()

# Survival count by passenger class
sns.countplot(data=df, x='pclass', hue='survived')
plt.title("Survival by Passenger Class")
plt.show()

# KDE plot of Age by Survival
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='age', hue='survived', fill=True)
plt.title("Age Distribution by Survival")
plt.show()
fig = px.histogram(df, x='age', color='survived', nbins=30,
                   marginal='box', title="Age Distribution with Survival")
fig.show()
```
